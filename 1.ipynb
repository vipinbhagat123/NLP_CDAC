# Ambiguity in Natural Language Processing (NLP)

Ambiguity is when a word, phrase, or sentence has multiple possible meanings. It's a key challenge for machines processing human language.

## Types of Ambiguity:

1.  **Lexical Ambiguity (Word-level)**
    * **Description:** A single word has multiple meanings.
    * **Subtypes:**
        * **Homonymy:** Words spelled/pronounced identically but with unrelated meanings.
            * **Example:** "bank" (financial institution) vs. "bank" (river edge)
        * **Polysemy:** Words with multiple *related* meanings.
            * **Example:** "head" (body part) vs. "head" (leader)
    * **Illustration:** "He found a **match**." (fire starter OR competition)

2.  **Syntactic Ambiguity (Structural)**
    * **Description:** A sentence can be grammatically parsed in more than one way, leading to different interpretations of its structure.
    * **Illustration:** "I saw the man with the telescope."
        * *Interpretation 1:* (I used a telescope to see the man)
        * *Interpretation 2:* (The man was holding a telescope)

3.  **Semantic Ambiguity (Meaning-level)**
    * **Description:** The overall meaning of a sentence or phrase is unclear, even if individual words are unambiguous.
    * **Illustration:** "The chicken is ready to eat."
        * *Interpretation 1:* (The chicken is cooked and ready to be consumed)
        * *Interpretation 2:* (The chicken animal is hungry and ready to consume food)

4.  **Referential Ambiguity (Anaphoric)**
    * **Description:** A pronoun or noun phrase could refer to more than one previously mentioned entity.
    * **Illustration:** "Alice told Jane that **she** would win."
        * *Interpretation 1:* ("she" refers to Alice)
        * *Interpretation 2:* ("she" refers to Jane)

5.  **Pragmatic Ambiguity**
    * **Description:** Uncertainty about the speaker's *intended* meaning, which goes beyond the literal words and relies on context, shared knowledge, or social conventions.
    * **Illustration:** "Can you pass the salt?"
        * *Literal Meaning:* (Are you physically capable of passing the salt?)
        * *Intended Meaning:* (Please pass the salt.)

## Resolving Ambiguity in NLP:

NLP systems employ various techniques:
* **Contextual Analysis:** Using surrounding words/sentences.
* **Word Sense Disambiguation (WSD):** Identifying the correct meaning (sense) of a word.
* **Syntactic Parsing:** Analyzing the grammatical structure.
* **Coreference Resolution:** Linking references to the same entity.
* **World Knowledge/Common Sense Reasoning:** Increasingly, large language models (LLMs) leverage learned patterns from vast datasets.
